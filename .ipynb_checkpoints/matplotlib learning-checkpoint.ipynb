```python
x = np.array([1, 1, 5, 5, 1])
y = np.array([1, 5, 5, 1, 1])
plt.plot(x, y)
plt.show()
```

```python
y = np.arange(0, 5, 1) # [0, 1, 2, 3, 4]
x = np.array([a*a for a in y]) # [0, 1, 4, 9, 16]
plt.plot(x, y)
plt.grid() # сетка
plt.show()
```

```python
y = np.arange(0, 5, 1)
x = np.array([a*a for a in y])
y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]
plt.plot(x, y, x2, y2) # ~ plt.plot(x, y); plt.plot(x2, y2)
plt.show()
```

```python
# Изменение стилей линий у графиков

#'-' Непрерывная линия (используется по умолчанию)
#
#'--' Штриховая линия
#
#'-.' Штрихпунктирная линия
#
#':' Пунктирная линия
#
#'None' или ' ' Без рисования линии


y = np.array([1, 2, -6, 0, 4])
plt.plot(y, '--')
plt.show()
```

```python
# Изменение цвета линий графиков

# <p align=center>{'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}

y = np.arange(0, 5, 1)
x = np.array([a*a for a in y])
y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]

lines = plt.plot(x, y, '--g', x2, y2, ':m')

# или в plt.plot() указать парамектр с or color:
#color='r'
#color='#0000CC' HEX
#с = (0, 0, 0) RGB
#с = (0, 0, 0, 0.5) RGBA
```

```python
# Изменение маркеров точек у графиков

plt.plot(x2, y2, ':o') # o - тип маркера
```

```python
lines = plt.plot(x, y, '-go', x2, y2, 's:m', markerfacecolor='w')
# или параметр marker:
#marker='d' ромб
```

```python
# Заливка областей графика

# fill_between(x,  y1,  y2=0,  where=None,  interpolate=False,  step=None,  *, data=None, **kwargs)

x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y)
plt.show()
```

```python
plt.fill_between(x, y, 0.5) # y2 = 0.5
```

```python
# дополнительно еще сделать ограничение на выбор заливаемого региона, когда y < 0
plt.fill_between(x, y, 0.5, where=(y < 0))
plt.show()
```

```python
plt.fill_between(x, y, where=(y < 0), color='r', alpha=0.5)
plt.fill_between(x, y, where=(y > 0), color='g', alpha=0.5)
```

```python
#3 Отображение нескольких координатных осей (Axes) в одном окне (Figure)

# subplot(nrows, ncols, index)

plt.subplot(1, 3, 1)
plt.plot(np.random.random(10))

plt.subplot(1, 3, 2)
plt.plot(np.random.random(10))

plt.subplot(1, 3, 3)
plt.plot(np.random.random(10))

plt.grid() 
# сетка отобразилась для последнего графика - для каждого графика указывать сетку
# Дело в том, что функция grid() сработала для активного объекта Axes
plt.show()
```

```python
ax1 = plt.subplot(1, 3, 1)
plt.plot(np.random.random(10))

ax2 = plt.subplot(1, 3, 2)
plt.plot(np.random.random(10))

ax3 = plt.subplot(1, 3, 3)
plt.plot(np.random.random(10))

ax1.grid()
ax2.grid()
ax3.grid()
```

```python
ax1 = plt.subplot(2, 3, 1)
plt.plot(np.random.random(10))

ax2 = plt.subplot(2, 3, 2)
plt.plot(np.random.random(10))

ax3 = plt.subplot(2, 3, 3)
plt.plot(np.random.random(10))

ax4 = plt.subplot(2, 1, 2)
plt.plot(np.random.random(10))
```

```python
# вместо трех чисел можно указывать трехзначное число

ax1 = plt.subplot(131)
plt.plot(np.random.random(10))

ax2 = plt.subplot(132)
plt.plot(np.random.random(10))

ax3 = plt.subplot(133)
plt.plot(np.random.random(10))
```

```python
# Функция subplots()

# Однако, с помощью функции subplot() не очень удобно оперировать большим числом координатных осей в пределах одной формы. 
#Для этого лучше подходит другая похожая функция:

# subplots(nrows, ncols)

# которая на выходе дает: 
# 1. ссылку на фигуру
# 2. список координатных осей:

f, ax = plt.subplots(2, 2) # 2x2

# Причем, оси здесь создаются сразу в отличие от функции subplot(). 
#Далее, мы можем взять любой объект Axes, чтобы работать с нужным нам полем, например, так:

ax[0, 0].plot(np.arange(0, 5, 0.2)) # ax[0][0]
ax[0, 0].grid()

ax[0, 1].plot(np.arange(5, 0, -0.2))
ax[0, 1].grid()
plt.show()
```

```python
# Объект Figure

# используя ссылку на объект Figure, можно устанавливать те или иные свойства для всего окна:

f.set_size_inches(7, 4) # размер 7 x 4 дюймов
f.set_facecolor('#eee') # светло-серый фон

# Объект Figure можно создавать и независимо с помощью класса Figure, например, так

fig = plt.figure(figsize=(7, 4))

#В результате, у нас появятся два окна для отображения графиков. 
#Это произошло по той причине, что функция subplots() автоматически создает новое окно. 
#Если нам нужно на ранее созданной фигуре отображать графики, то мы должны сразу после создания записать функцию plot():

plt.plot(np.arange(0, 5, 0.2))

# add_axes = [pos_x, pos_y, width, height]
```

```python
# другой способ

fig = plt.figure(figsize=(7, 4))
ax1 = fig.add_subplot(1, 3, 1)
ax1.plot(np.arange(0, 5, 0.2))
```

```python
# Компоновка графиков с помощью GridSpec

# Для более сложных элементов компоновки хорошо подходит класс GridSpec из ветки

fig = plt.figure(figsize=(7, 4))
gs = GridSpec(ncols=3, nrows=2, figure=fig)

ax1 = plt.subplot(gs[0, 0])
ax1.plot(np.arange(0, 5, 0.2))

ax2 = fig.add_subplot(gs[1, 0:2])
ax2.plot(np.random.random(10))

ax3 = fig.add_subplot(gs[:, 2])
ax3.plot(np.random.random(10))

#Причем, обратите внимание, первую ось я специально сформировал с помощью функции subplot(), а остальные две – с помощью метода add_subplot(). 
#То есть, можно использовать любой способ формирования полей для вывода графиков.
```

```python
ws = [1, 2, 5]
hs = [2, 0.5]

gs = GridSpec(ncols=3, nrows=2, figure=fig, width_ratios=ws, height_ratios=hs)
```
